## Perf. factors

Summarize performance of selected algos while varying some benchid parameters.  Plots generally show 95% CIs and lines drawn between medians for each algo wrt some performance measure (1000 point sample).  Continuous parameters are sometimes discretized as needed.

Over all, it seems difficult to distinguish the performance of different algos using AUC, compared to the other performance measures.

```python
measures = ['APR','AUC','RPrec','p@0.1%','p@5%']
```

```python
plt.figure(figsize=(15,4))
ax = plt.gca()
scr_met[measures].reset_index().drop(labels='benchid',axis=1).groupby('algo').agg(np.average,axis=0).sort('p@0.1%',ascending=False).plot(kind='bar', ax=ax)
plt.yscale('log')
plt.legend(ncol=1,bbox_to_anchor=(1.01, 1), loc=2)
plt.show()
```

### num. feats

As the number of features in the dataset rises
 + Algo performance generally decreases, although that pattern is clearer in harder benchids.

```python
scr_disp = scr_sel2.copy()
scr_disp['features'] = pd.qcut(scr_disp.avgFeatures, 7)
for m in measures:
    print 'Measure: %s'%m
    sns.factorplot("features", m, hue='algo', 
                   col_wrap=2,
                   x_order=scr_disp.features.unique(),
                   col="diff", data=scr_disp, size=4, aspect=1.5,
                   col_order=['easy','medium','hard','veryhard'],
                   palette=sns.color_palette("Set2", n_colors=len(alg_sel)))
    plt.show()
```

### num. inst.

As the size of the dataset increases
 + No over-all trend in performance.

```python
scr_disp = scr_sel2.copy()
scr_disp['instCnt'] = pd.qcut(scr_disp.instCnt, 7)
for m in measures:
    print 'Measure: %s'%m
    sns.factorplot("instCnt", m, hue='algo', 
                   col_wrap=2,
                   x_order=scr_disp.instCnt.unique(),
                   col="diff", data=scr_disp, size=4, aspect=1.75,
                   col_order=['easy','medium','hard','veryhard'],
                   palette=sns.color_palette("Set2", n_colors=len(alg_sel)))
    plt.show()
```

### num. anom.

As the number of anomalies in a benchid increases
 + Precision-based measures general improve.

```python
scr_disp = scr_sel2.copy()
scr_disp['numAnomalies'] = pd.qcut(scr_disp.numAnomalies, 7)
for m in measures:
    sns.factorplot("numAnomalies", m, hue='algo', 
                   col_wrap=2,
                   x_order=scr_disp.numAnomalies.unique(),
                   col="diff", data=scr_disp, size=4, aspect=1.5,
                   col_order=['easy','medium','hard','veryhard'],
                   palette=sns.color_palette("Set2", n_colors=len(alg_sel)))
    plt.show()
```

### relFreq

As the number of anomalies in a benchid increases relative to the total number of points
 + Precision-based measures general improve.
 + AUC decreases

```python
for m in measures:
    sns.factorplot("relFreq", m, hue='algo', 
                   #row="dataset", 
                   col_wrap=2,
                   col="diff", data=scr_sel2, size=4, aspect=1.5,
                   col_order=['easy','medium','hard','veryhard'],
                   palette=sns.color_palette("Set2", n_colors=len(alg_sel)))
    plt.show()
```

### difficulty

As the difficulty category of benchids increases
 + Performance generally decreases.

```python
scr_disp = scr_sel2.copy()
scr_disp['instCnt'] = pd.qcut(scr_disp.instCnt, 4)
for m in measures:
    print 'Measure: %s'%m
    sns.factorplot("diff", m, hue='algo', 
                   col_wrap=2,
                   x_order=['easy','medium','hard','veryhard'],
                   col="instCnt", data=scr_disp, size=4, aspect=1.25,
                   col_order=scr_disp.instCnt.unique(),
                   palette=sns.color_palette("Set2", n_colors=len(alg_sel)))
    plt.show()
```
